$f(x)=1260+117x-9x^2$

```python
import pandas as pd
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
def f(x):
    return 1260 + 117*x - 9*x**2
f(x)
```

```python
df = sp.diff(f(x), x)
df
```

```python
pc = sp.solve(f(x), x)
pc
```

```python
xn = np.linspace(-100 , 100, 500)#(start, stop, n)

yn = f(xn)

plt.figure()
plt.plot(xn, yn)

plt.plot(pc[0], f(pc[0]), "+r")

```

```python
def g(x):
    return 1260 + 117*x - 9*x**2
dg = sp.diff (g(x), x)
pc = sp.solve(dg, x)

pc
```

```python
yn = g(xn)

plt.figure()
plt.plot(xn, yn)
plt.plot(pc[0], g(pc[0]), "og")
```
